```python
genre_global_sales = df.filter(items=["Genre","Global_Sales"]).groupby(by="Genre").sum().sort_values("Global_Sales")
```

```python
genre_global_sales
```

# Video Game Global Sales in Genre

```python
genre_global_sales.plot(kind="bar",title="Video Game Global Sales in Genre",color="yellow")
```

```python
year_global_sales = df.filter(items=["Year","Global_Sales"]).groupby(by="Year").sum()
```
